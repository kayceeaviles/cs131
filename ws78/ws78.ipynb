```python
import time
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("CSV Run Time").getOrCreate()

# csv files are organized by size
csv_files = [
    ("data_sample3.csv", "gs://dataproc-staging-us-central1-325900183882-rh5w8gsb/datasets/data_sample3.csv"),
    ("data_sample2.csv", "gs://dataproc-staging-us-central1-325900183882-rh5w8gsb/datasets/data_sample2.csv"),
    ("data_sample.csv", "gs://dataproc-staging-us-central1-325900183882-rh5w8gsb/datasets/data_sample.csv"),
    ("data_sample4.csv", "gs://dataproc-staging-us-central1-325900183882-rh5w8gsb/datasets/data_sample4.csv"),
    ("data_sample5.csv", "gs://dataproc-staging-us-central1-325900183882-rh5w8gsb/datasets/data_sample5.csv"),
]

results = []

for file_name, gcs_path in csv_files:
    start_time = time.time()
    df = spark.read.csv(gcs_path, header=True, inferSchema=True)
    df.count()
    end_time = time.time()
    load_time = end_time - start_time
    results.append((file_name, load_time))
    print(f"{file_name}: {load_time:.2f} seconds.")

load_times = [x[1] for x in results]
file_sizes = [4, 5, 7, 8, 12]

plt.figure(figsize=(10, 6))
plt.scatter(file_sizes, load_times, color='pink', marker='o')
plt.plot(file_sizes, load_times, color='pink', linestyle='-', marker='o')

plt.title("File Size vs Load Time")
plt.xlabel("File Size (KB)")
plt.ylabel("Time (seconds)")
plt.grid(True)

plt.show()
```
